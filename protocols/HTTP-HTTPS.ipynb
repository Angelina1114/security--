## 1. HTTP 基礎與協議堆疊

### HTTP 依賴的其他協議

HTTP 不是獨立運作的，它需要依靠其他協議：

| 協議層 | 協議 | 作用 |
|--------|------|------|
| **應用層** | **HTTP/HTTPS** | 網頁內容傳輸 |
| **傳輸層** | **TCP** | 可靠的資料傳輸 |
| **網路層** | **IP** | 封包路由 |
| **資料鏈路層** | **Ethernet/WiFi** | 實體網路傳輸 |

### 相關協議說明

**DNS（域名解析）**：
- 將網址轉換為 IP 位址
- `example.com` → `93.184.216.34`

**TCP（傳輸控制協議）**：
- 確保資料完整傳輸
- HTTP 使用 TCP 80 埠，HTTPS 使用 TCP 443 埠

**TLS/SSL（安全協議）**：
- HTTPS 的加密層
- 在 HTTP 和 TCP 之間提供加密

## 1.5 URL 網址組成結構

### URL 的完整格式
```
https://www.example.com:443/path/to/page?name=value&id=123#section
  ↑         ↑        ↑    ↑           ↑                    ↑
協議      域名      埠號   路徑        查詢參數              錨點
```

### 各部分詳細說明

| 部分 | 範例 | 說明 | 必要性 |
|------|------|------|--------|
| **協議** | `https://` | 使用的傳輸協議 | 必要 |
| **域名** | `www.example.com` | 伺服器的網域名稱 | 必要 |
| **埠號** | `:443` | 服務埠號 | 可選 (HTTP=80, HTTPS=443) |
| **路徑** | `/path/to/page` | 伺服器上的檔案路徑 | 可選 (預設為 /) |
| **查詢參數** | `?name=value&id=123` | 傳遞給伺服器的參數 | 可選 |
| **錨點** | `#section` | 頁面內的特定位置 | 可選 (不會傳給伺服器) |

### 實際範例分析

#### 搜尋網址範例
```
https://www.google.com/search?q=網路安全&hl=zh-TW
```
- **協議**：`https://` (安全連線)
- **域名**：`www.google.com` (Google 搜尋引擎)
- **路徑**：`/search` (搜尋功能)
- **參數**：`q=網路安全` (搜尋關鍵字), `hl=zh-TW` (語言設定)

#### 購物網站範例
```
https://shop.example.com:8080/products/laptops?brand=apple&price=1000-2000#reviews
```
- **協議**：`https://` (安全購物)
- **域名**：`shop.example.com` (購物網站)
- **埠號**：`:8080` (非標準埠)
- **路徑**：`/products/laptops` (筆電分類)
- **參數**：`brand=apple&price=1000-2000` (品牌和價格篩選)
- **錨點**：`#reviews` (跳到評論區塊)

### URL 與 HTTP 請求的關係
- **域名** → DNS 解析為 IP 位址
- **協議 + 埠號** → 決定連線方式
- **路徑 + 參數** → 包含在 HTTP 請求中

````python
# URL 分析實作
echo "=== URL 組成分析 ==="

# 使用 curl 顯示請求的詳細資訊
echo "分析這個 URL: https://httpbin.org/get?name=test&id=123"
curl -v "https://httpbin.org/get?name=test&id=123" 2>&1 | grep -E "(GET|Host:)" 

echo -e "\n檢查不同協議的預設埠號:"
echo "HTTP (埠號 80):"
curl -I http://example.com | head -1
echo "HTTPS (埠號 443):"  
curl -I https://example.com | head -1

echo -e "\n顯示查詢參數如何傳遞:"
curl -s "https://httpbin.org/get?user=student&course=security" | grep -A5 '"args"'
````

## 2. HTTP 完整運作流程

### 當你在瀏覽器輸入 `http://example.com` 時發生什麼？

#### 步驟 1：DNS 域名解析
```
用戶輸入：http://example.com
瀏覽器問：example.com 的 IP 位址是什麼？
DNS 伺服器答：93.184.216.34
```

#### 步驟 2：建立 TCP 連線
```
瀏覽器 → 伺服器 (93.184.216.34:80)
三次握手：SYN → SYN-ACK → ACK
連線建立成功
```

#### 步驟 3：發送 HTTP 請求
```
GET / HTTP/1.1
Host: example.com
User-Agent: Mozilla/5.0 ...
Accept: text/html,application/xhtml+xml
```

#### 步驟 4：伺服器處理請求
```
伺服器接收請求 → 找到對應資源 → 準備回應
```

#### 步驟 5：伺服器回傳 HTTP 回應
```
HTTP/1.1 200 OK
Content-Type: text/html
Content-Length: 1270
Server: Apache/2.4.41

<html>
<head><title>Example</title></head>
<body>歡迎來到範例網站</body>
</html>
```

#### 步驟 6：瀏覽器處理回應
```
接收 HTML → 解析內容 → 渲染網頁 → 顯示給用戶
```

#### 步驟 7：關閉連線
```
資料傳輸完成 → TCP 四次揮手 → 連線關閉
```

### 完整流程圖
```
用戶 → 瀏覽器 → DNS → TCP → HTTP請求 → 伺服器
                ↑                            ↓
              顯示網頁 ← 渲染 ← HTML ← HTTP回應
```

````python
# 模擬 HTTP 請求流程
echo "=== HTTP 運作流程示範 ==="

echo "步驟 1: DNS 查詢"
nslookup example.com | grep "Address" | tail -2

echo -e "\n步驟 2-7: 完整 HTTP 請求 (含詳細資訊)"
curl -v http://example.com 2>&1 | head -20

echo -e "\n顯示完整的 HTTP 標頭："
curl -I http://example.com

echo -e "\n檢查連線時間："
curl -w "DNS解析: %{time_namelookup}s\n建立連線: %{time_connect}s\n總時間: %{time_total}s\n" -o /dev/null -s http://example.com
````

## 3. HTTP vs HTTPS 流程差異

### HTTP 流程 (不安全)
```
瀏覽器 ←--明文傳輸--> 伺服器
   ↑                    ↓
 容易被              資料可見
 竊聽攔截            易被修改
```

### HTTPS 流程 (安全)
```
瀏覽器 ←--TLS加密--> 伺服器
   ↑                   ↓
 加密保護            身份驗證
 無法竊聽            資料完整
```

### 主要差異表格

| 特性 | HTTP | HTTPS |
|------|------|-------|
| **安全性** | ❌ 明文傳輸 | ✅ 加密傳輸 |
| **埠號** | 80 | 443 |
| **額外步驟** | 無 | TLS 握手 |
| **瀏覽器顯示** | ⚠️ 不安全 | 🔒 安全鎖 |
| **用途** | 一般資訊 | 敏感資料 |

### 重要狀態碼
- **200 OK**：請求成功
- **301/302**：重新導向到新位址
- **404 Not Found**：找不到頁面
- **500 Internal Server Error**：伺服器錯誤

### HTTPS 額外的 TLS 握手過程

當使用 HTTPS 時，在 HTTP 通訊前還有額外步驟：

#### TLS 握手流程
```
步驟 1: 客戶端 → 伺服器
"Hello, 我支援這些加密方式"

步驟 2: 伺服器 → 客戶端  
"Hello, 我選擇這個加密方式，這是我的憑證"

步驟 3: 客戶端驗證憑證
檢查憑證是否可信任 → 產生加密金鑰

步驟 4: 雙方確認加密金鑰
建立安全通道 → 開始 HTTP 通訊
```

#### 完整 HTTPS 流程
```
DNS解析 → TCP連線 → TLS握手 → HTTP請求 → HTTP回應 → 關閉連線
                      ↑
                   額外的安全步驟
```

## 4. HTTP 請求方法詳解

### 常見 HTTP 方法

| 方法 | 用途 | 特性 | 常見用法 |
|------|------|------|----------|
| **GET** | 取得資料 | 安全、冪等、可快取 | 瀏覽網頁、下載檔案 |
| **POST** | 提交資料 | 不安全、非冪等 | 表單送出、上傳檔案 |
| **PUT** | 更新資料 | 不安全、冪等 | 完整更新資源 |
| **PATCH** | 部分更新 | 不安全、非冪等 | 修改部分內容 |
| **DELETE** | 刪除資料 | 不安全、冪等 | 刪除資源 |
| **HEAD** | 取得標頭 | 安全、冪等、可快取 | 檢查資源是否存在 |
| **OPTIONS** | 查詢方法 | 安全、冪等 | CORS 預檢請求 |

### GET vs POST 詳細比較

| 特性 | GET | POST |
|------|-----|------|
| **資料位置** | URL 參數 | 請求主體 |
| **資料長度** | 有限制 (約2KB) | 無限制 |
| **安全性** | 參數在 URL 可見 | 參數在主體內 |
| **快取** | 可被快取 | 通常不快取 |
| **書籤** | 可加入書籤 | 無法加書籤 |
| **重複執行** | 安全 | 可能造成問題 |

### 實際請求格式範例

#### GET 請求
```
GET /search?q=網路安全 HTTP/1.1
Host: example.com
User-Agent: Mozilla/5.0...
Accept: text/html,application/json
```

#### POST 請求
```
POST /login HTTP/1.1
Host: example.com
Content-Type: application/x-www-form-urlencoded
Content-Length: 27

username=user&password=pass
```

````python
# HTTP 請求方法實際操作

# GET 請求 (預設方法)
curl -X GET https://httpbin.org/get

# POST 請求發送資料
curl -X POST -d "name=test&value=123" https://httpbin.org/post

# PUT 請求更新資料  
curl -X PUT -d '{"key":"value"}' -H "Content-Type: application/json" https://httpbin.org/put

# DELETE 請求
curl -X DELETE https://httpbin.org/delete

# HEAD 請求 (只取標頭)
curl -I -X HEAD https://httpbin.org/get

# OPTIONS 請求 (查詢支援的方法)
curl -X OPTIONS https://httpbin.org/
````

````python
# 基本 HTTP 檢測指令
echo "=== HTTP/HTTPS 基本檢測 ==="

echo "1. 檢查網站狀態："
curl -I https://www.google.com

echo -e "\n2. 比較 HTTP vs HTTPS："
curl -I http://example.com
curl -I https://example.com

echo -e "\n3. 檢查回應時間："
time curl -s https://www.github.com > /dev/null
````

## 3. 重要 HTTP 標頭

### 基本標頭
- **Status**：狀態碼（200、404、500等）
- **Server**：伺服器軟體
- **Content-Type**：內容類型（html、json等）

### 安全標頭
- **Strict-Transport-Security**：強制使用 HTTPS
- **Content-Security-Policy**：防止惡意腳本
- **X-Frame-Options**：防止頁面被嵌入

## 4. 安全重點

### 🚨 HTTP 風險
- **明文傳輸**：密碼可被竊聽
- **內容竄改**：資料可能被修改
- **身份偽造**：難以確認網站真實性

### 🛡️ HTTPS 保護
- **加密傳輸**：資料被加密保護
- **身份驗證**：確認網站身份
- **完整性檢查**：防止內容被竄改
